```python
data = open("AoC2021-24.input").read()

code = [x.split() for x in data.strip().splitlines()]

regs = {'w': 0, 'x': 0, 'y': 0, 'z': 0}

inpdata = [0] * 14

def zeregs():
    global regs
    regs = {'w': 0, 'x': 0, 'y': 0, 'z': 0}

def makeinp(x):
    global inpdata
    return list(map(int, str(x)))

def val(z):
    if z in regs:
        return regs[z]
    else:
        return int(z)

def vals(cmd):
    return val(cmd[1]), val(cmd[2])

def inp(cmd):
    regs[cmd[1]] = inpdata.pop(0) 

def add(cmd):
    a, b = vals(cmd)
    regs[cmd[1]] = a + b

def sub(cmd):
    a, b = vals(cmd)
    regs[cmd[1]] = a - b

def mul(cmd):
    a, b = vals(cmd)
    regs[cmd[1]] = a * b

def div(cmd):
    a, b = vals(cmd)
    regs[cmd[1]] = a // b
    # c = abs(a) // abs(b)
    # if a * b < 0:
    #     c = -c
    # regs[cmd[1]] = c

def mod(cmd):
    a, b = vals(cmd)
    regs[cmd[1]] = a % b

def eql(cmd):
    a, b = vals(cmd)
    regs[cmd[1]] = int(a == b)

_ops = {
    'add': add,
    'sub': sub,
    'mul': mul,
    'div': div,
    'mod': mod,
    'eql': eql,
    'inp': inp,
}

def xcompile(code):
    ret = """def run(input):
    w = x = y = z = 0
    """
    for c in code:
        if c[0] == 'add':
            ret += f'{c[1]} += {c[2]}'
        elif c[0] == 'sub':
            ret += f'{c[1]} -= {c[2]}'
        elif c[0] == 'mul':
            ret += f'{c[1]} *= {c[2]}'
        elif c[0] == 'div':
            ret += f'{c[1]} //= {c[2]}'
        elif c[0] == 'mod':
            ret += f'{c[1]} %= {c[2]}'
        elif c[0] == 'eql':
            ret += f'{c[1]} = int({c[1]} == {c[2]})'
        elif c[0] == 'inp':
            ret += f'{c[1]} = input.pop(0)'
        ret += '\n    '
    ret += """return w, x, y, z\n"""  
    return ret

        



inpdata = makeinp(11111111111111)

print(*[f"{k}={v}" for k, v in regs.items()])
for cmd in code:
    _ops[cmd[0]](cmd)
print(*[f"{k}={v}" for k, v in regs.items()])




```
